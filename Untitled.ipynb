### Modelo del problema:

Se tiene un tazón cuadrado al que se le quito en la parte superior lo que correspondería al volumen de un paraboloide. El volumen de este tazón esta comprendido dentro de las inecuaciones:

$$z \leq \sqrt{1+.5x^2+.5y^2}$$
$$\vert x \vert \leq 8$$
$$\vert y \vert \leq 8$$

De estas últimas dos inequaciones se puede ver como la base de la caja mide $16 \times 16$ unidades$^{2}$. Mientras que la altura esta comprendida va de $z=0$ hasta $z=\sqrt{1+32+32}=\sqrt{65}$.

Viendosé así la superficie del tazón:

```python
# Importacion
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np

#Definición de la formula de la superficie
def f(x,y):
    return np.sqrt(1+.5*x*x+.5*y*y)

#Grafica
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')
    #Coordeandas
xnum = np.linspace(-8, 8, 100)
ynum = np.linspace(-8, 8, 100)

X,Y = np.meshgrid(xnum, ynum)
ax.plot_surface(X,Y,f(X,Y));
```
